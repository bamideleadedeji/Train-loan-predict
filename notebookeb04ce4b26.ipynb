```python
train_data['Loan_Status'].replace(1,'Yes', inplace = True)
train_data['Loan_Status'].replace(0, 'No', inplace = True)
```

```python
train_data
```

```python
train_data.drop('Loan_ID', axis = 1, inplace = True)
```

```python
train_data['Gender'] = np.where((train_data['Gender'] == 'Yes'), 1, 0)
```

```python
train_data['Married'] = np.where((train_data['Married'] == 'Yes'), 1, 0)
```

```python
train_data['Education'] = np.where((train_data['Education'] == 'Yes'), 1, 0)
```

```python
train_data['Self_Employed'] = np.where((train_data['Self_Employed'] == 'Yes'), 1, 0)
```

```python
train_data['Property_Area'] = np.where((train_data['Property_Area'] == 'Yes'), 1, 0)
```

```python
train_data['Property_Area'] = np.where((train_data['Property_Area'] == 'Yes'), 1, 0)
```

```python
train_data['Dependents'] = np.where((train_data['Dependents'] == 'Yes'), 1, 0)
```

```python
le = LabelEncoder()
train_data['Loan_Status'] = le.fit_transform(train_data['Loan_Status'])
```

```python
X = train_data.drop('Loan_Status', axis = 1)
y = train_data['Loan_Status']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
dtree=DecisionTreeClassifier()
model=dtree.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
logreg=LogisticRegression()
model=logreg.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ens=RandomForestClassifier()
model=ens.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
nai=GaussianNB()
model=nai.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
k=KNeighborsClassifier()
model=k.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
